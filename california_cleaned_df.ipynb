```python
#Our data frame with only CA data
cali_df = accidents_df.loc[(accidents_df["State"] == "CA"), :]
```

```python
cleaned_df = cali_df[['ID', 'Severity', 'Start_Time', 'Start_Lat', 'Start_Lng', 'Description',
                      'City', 'County', 'State', 'Zipcode', 'Country', 'Timezone',
                      'Weather_Timestamp', 'Temperature(F)', 'Wind_Chill(F)', 'Humidity(%)',
                      'Pressure(in)', 'Visibility(mi)', 'Wind_Direction', 'Wind_Speed(mph)', 
                      'Precipitation(in)', 'Weather_Condition', 'Crossing', 'Junction', 
                      'No_Exit', 'Station', 'Stop', 'Traffic_Signal', 'Sunrise_Sunset']]
```

```python
#Our Main df with only CA data and neccessary columns
cleaned_df.head()
```

```python
#10 days before late oct early nov
#10 days after late feb early march
cleaned_df[["date", "time"]] = cleaned_df["Start_Time"].str.split(" ", expand = True)
cleaned_df[["year", "month", "day"]] = cleaned_df["date"].str.split("-", expand = True)
cleaned_df[["year", "month", "day"]] = cleaned_df[["year", "month", "day"]].astype(int)
cleaned_df.head()
```

```python
#Filter Data for months involved with Day Light Savings
final_df = cleaned_df.loc[(cleaned_df["month"] == 2) | 
               (cleaned_df["month"] == 3) |
               (cleaned_df["month"] == 10) |
              (cleaned_df["month"] == 11), :]
```

```python
#Arizona Dataframe
#Filter out unnecessary columns
#filter only for Arizona data
start_df = accidents_df.loc[(accidents_df["State"] == "AZ"), :]
ari_df = start_df[['ID', 'Severity', 'Start_Time', 'Start_Lat', 'Start_Lng', 'Description',
                      'City', 'County', 'State', 'Zipcode', 'Country', 'Timezone',
                      'Weather_Timestamp', 'Temperature(F)', 'Wind_Chill(F)', 'Humidity(%)',
                      'Pressure(in)', 'Visibility(mi)', 'Wind_Direction', 'Wind_Speed(mph)', 
                      'Precipitation(in)', 'Weather_Condition', 'Crossing', 'Junction', 
                      'No_Exit', 'Station', 'Stop', 'Traffic_Signal', 'Sunrise_Sunset']]
```

```python
#Create a new columns for data and time to be spilt up
#10 days before late oct early nov
#10 days after late feb early march
ari_df[["date", "time"]] = ari_df["Start_Time"].str.split(" ", expand = True)
ari_df[["year", "month", "day"]] = ari_df["date"].str.split("-", expand = True)
ari_df[["year", "month", "day"]] = ari_df[["year", "month", "day"]].astype(int)
ari_df.head()
```

```python
#Filter Data for months involved with Day Light Savings
arizona_df = ari_df.loc[(ari_df["month"] == 2) | 
               (ari_df["month"] == 3) |
               (ari_df["month"] == 10) |
              (ari_df["month"] == 11), :]
```
